```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data["sex"].replace({"female":0,"male":1},inplace=True)
```

```python
data["smoker"].replace({"no":0,"yes":1},inplace=True)
```

```python
data["region"].value_counts()
```

```python
data["region"].replace({"southeast":1,"southwest":2,"northwest":3,"northeast":4},inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2

def r2_score(y_true, y_pred):
    ss_res = K.sum(K.square(y_true - y_pred))
    ss_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return 1 - ss_res / (ss_tot + K.epsilon())

model = Sequential()
model.add(Dense(128, activation="relu", input_dim=6, kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(64, activation="relu", kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(32, activation="relu", kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dropout(0.1))

model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1))

model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='mean_squared_error',
    metrics=[r2_score, 'mae']
)

early_stopping = EarlyStopping(monitor='val_r2_score', patience=10, mode='max', restore_best_weights=True)

history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping]
)
```

```python
import pickle

pickle.dump(scaler,open("charges_scaler.sav","wb"))

model.save("charges_model.h5")
```
